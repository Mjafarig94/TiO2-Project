```python
MaxGlobal=0
MaxGlobal_Score=0
MaxGlobal_Score_train=0

scores = {'CV': [], 'R2': [],'R2_train': []}

for j in range(5000,5051):
    from sklearn.model_selection import cross_val_score
    from sklearn.ensemble import GradientBoostingRegressor
    import numpy as np
    import pandas as pd
    import pickle
    from sklearn import metrics
    from sklearn import preprocessing
    from sklearn.model_selection import train_test_split
    from sklearn.model_selection import cross_val_predict
    import matplotlib.pyplot as plt
    #from neupy.algorithms import RBFKMeans
    #from neupy.algorithms import GBR
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.metrics import r2_score
    from math import sqrt
    from sklearn.metrics import mean_absolute_error as mae
    from sklearn import svm
    from sklearn.neural_network import MLPRegressor
    from sklearn.neural_network import MLPClassifier
    from sklearn.linear_model import LogisticRegression
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.ensemble import AdaBoostRegressor
    from sklearn import linear_model
    import seaborn as sn
    import joblib
    import pandas as pd
    from sklearn.linear_model import LogisticRegression
    from sklearn.model_selection import GridSearchCV
    import warnings
    warnings.filterwarnings('ignore', category=FutureWarning)
    warnings.filterwarnings('ignore', category=DeprecationWarning)





    file_name = 'CA'
    dataset = pd.read_csv(file_name + '.csv', header=0)
    Test=dataset.loc[dataset['phi'] == 0.5]
    data= dataset[dataset.phi != 0.5]

    x1 = data.phi
    x2= data.ca
    X = np.column_stack((x1,x2))
    Y= np.asarray(data.uv).reshape(-1, 1)
    X_train, X_test,Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state=j,stratify=X[:,0])


    xscaler = preprocessing.StandardScaler()
    X_train_standard=xscaler.fit_transform(X_train)
    X_test_standard=xscaler.transform(X_test)

    yscaler = preprocessing.StandardScaler()
    Y_train_standard =yscaler.fit_transform(Y_train)
    Y_test_standard =yscaler.transform(Y_test)

    Y_train_standard=Y_train_standard.ravel()
    Y_test_standard =Y_test_standard.ravel() 
    
    X1 = Test.phi
    X2= Test.ca
    XX = np.column_stack((X1,X2))
    YY= np.asarray(Test.uv).reshape(-1, 1)
    X_scaled=xscaler.transform(XX)
    Y_scaled=yscaler.transform(YY)
    Y_scaled=Y_scaled.ravel()
   



    def print_results_max(results):
        print('BEST PARAMS: {}\n'.format(results.best_params_))
        MAxMean=0
        means = results.cv_results_['mean_test_score']
        stds = results.cv_results_['std_test_score']
        for mean, std, params in zip(means, stds, results.cv_results_['params']):
            if mean > MAxMean :
                MAxMean = mean
                print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
        return MAxMean






    GradientBoostingRegressor = GradientBoostingRegressor()
    parameters = {
            'loss' : ['ls', 'lad', 'huber', 'quantile'],
            'n_estimators' : [5],
            'learning_rate': [0.01,0.1,1],
            'subsample' : [1],
            'criterion' : [ 'mse', 'mae'],
        'min_samples_split': [3],
        'max_depth' :[4],
        'max_features': ['auto'],
        'random_state': [10]

    }
    print("iteration {0}".format(j))

    cv = GridSearchCV(GradientBoostingRegressor, parameters, cv=5)
    cv.fit(X_train_standard, Y_train_standard)

    Score= cv.best_estimator_.score(X_test_standard, Y_test_standard).round(4)
    MAxMean=print_results_max(cv)
    Score_train= cv.best_estimator_.score(X_train_standard, Y_train_standard).round(4)

    scores['CV'].append(MAxMean) 
    scores['R2'].append(Score)
    scores['R2_train'].append(Score_train)

    
    if MAxMean > MaxGlobal and Score>MaxGlobal_Score  and Score_train>MaxGlobal_Score_train:
        MaxGlobal=MAxMean
        MaxGlobal_Score= Score
        MaxGlobal_Score_train = Score_train

        BestFit=cv.best_estimator_
        GB_X_train_standard=X_train_standard
        GB_Y_train_standard=Y_train_standard
        GB_X_test_standard=X_test_standard
        GB_Y_test_standard=Y_test_standard
        
        GBR_Y_train_standard_realscale=yscaler.inverse_transform(Y_train_standard.reshape(-1, 1))
        GBR_Y_test_standard_realscale=yscaler.inverse_transform(Y_test_standard.reshape(-1, 1))
        ypredtrain=BestFit.predict(X_train_standard)
        ypredtest=BestFit.predict(X_test_standard)
        GBR_pred_Y_train_standard_realscale=yscaler.inverse_transform(ypredtrain.reshape(-1, 1))
        GBR_pred_Y_test_standard_realscale=yscaler.inverse_transform(ypredtest.reshape(-1, 1))
        np.save('GBR_Y_train_standard_realscale',GBR_Y_train_standard_realscale)
        np.save('GBR_Y_test_standard_realscale',GBR_Y_test_standard_realscale)
        np.save('GBR_pred_Y_train_standard_realscale',GBR_pred_Y_train_standard_realscale)
        np.save('GBR_pred_Y_test_standard_realscale',GBR_pred_Y_test_standard_realscale)
        
        Number_Of_Estimators = j
        np.save('GB_fullRun_X_train_standard',GB_X_train_standard)
        np.save('GB_fullRun_Y_train_standard',GB_Y_train_standard)
        np.save('GB_fullRun_X_test_standard',GB_X_test_standard)
        np.save('GB_fullRun_Y_test_standard',GB_Y_test_standard)
        joblib.dump(BestFit, 'Gradboost_fullRun_model.pkl')
        print("MaxGlobal {0}".format(MaxGlobal))
        print("Number_Of_Estimators {0}".format(Number_Of_Estimators))
        print("MaxGlobal_Score {0}".format(MaxGlobal_Score))
        Xscaler=xscaler
        Yscaler=yscaler
        ypredTest=BestFit.predict(X_scaled)
        Testscore=r2_score(Y_scaled,ypredTest)

        
    
S=pd.DataFrame(scores) 
axis = S.boxplot(whis=(5,95))
axis.set_ylim([0,1]);
np.save('GB_CV',S.CV)
np.save('GB_R2',S.R2)
np.save('GB_R2_train',S.R2_train)

print("mean CV {0}".format(S.CV.mean()))
print("mean R2 {0}".format(S.R2.mean()))
print("mean R2_train {0}".format(S.R2_train.mean()))
print("Test Score {0}".format(Testscore))



print("Test 0.5 Score {0}".format(Testscore))
print("test r2 = {}".format(r2_score(GBR_Y_test_standard_realscale,GBR_pred_Y_test_standard_realscale)))
print("test RMSE: {0} and MAE: {1}".format(sqrt(np.mean((GBR_Y_test_standard_realscale - GBR_pred_Y_test_standard_realscale)**2)),mae(GBR_Y_test_standard_realscale, GBR_pred_Y_test_standard_realscale).round(3)))

print("train r2 = {}".format(r2_score(GBR_Y_train_standard_realscale,GBR_pred_Y_train_standard_realscale)))
print("train RMSE: {0} and MAE: {1}".format(sqrt(np.mean((GBR_Y_train_standard_realscale - GBR_pred_Y_train_standard_realscale)**2)),mae(GBR_Y_train_standard_realscale, GBR_pred_Y_train_standard_realscale).round(3)))
```
